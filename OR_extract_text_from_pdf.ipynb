```python
def parse_pdf(p):
    pdf_file = open(p, 'rb')
    read_pdf = PyPDF2.PdfFileReader(pdf_file)
    number_of_pages = read_pdf.getNumPages()
    page = read_pdf.getPage(0)
    page_content = page.extractText()
    length = [r for r in page_content.split('\n') if r.startswith('Length of protein sequence')]
    tm_number = [r for r in page_content.split('\n') if r.startswith('Number of modified TM regions')]
    
    return length[0].split()[-1], tm_number[0].split()[-1]
```

```python
d_length = {}
d_tm = {}

for file in pdf_files:
    p = email_output + file
    length, tm = parse_pdf(p)
    d_length[file.split('.')[0]] = length
    d_tm[file.split('.')[0]] = tm
```

```python
def read_fasta(fp):
        name, seq = None, []
        for line in fp:
            line = line.rstrip()
            if line.startswith(">"):
                if name: yield (name, ''.join(seq))
                name, seq = line, []
            else:
                seq.append(line)
        if name: yield (name, ''.join(seq))
       

d = defaultdict(list)    

with open(input_path + input_file) as rf:
    for name, seq in read_fasta(rf):
        parse_name = name.split('|')
        uniprot = parse_name[1]
        job_title = parse_name[2].split()[0]
        description = ' '.join(parse_name[2].split()[1:4])
        d[uniprot].extend([job_title, description, seq])
```
