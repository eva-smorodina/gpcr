```python
import os
import pandas as pd
from collections import defaultdict

from splinter import Browser # pip install splinter
```

```python
input_path = '/home/eva/Internships/Labs/Shuguang_Zhang/data/input/'
input_file = 'test_fasta.txt'
```

```python
def read_fasta(fp):
        name, seq = None, []
        for line in fp:
            line = line.rstrip()
            if line.startswith(">"):
                if name: yield (name, ''.join(seq))
                name, seq = line, []
            else:
                seq.append(line)
        if name: yield (name, ''.join(seq))
            

d = defaultdict(list)    

with open(input_path + input_file) as rf:
    for name, seq in read_fasta(rf):
        parse_name = name.split('|')
        uniprot = parse_name[1]
        job_title = parse_name[2].split()[0]
        description = ' '.join(parse_name[2].split()[1:4])
        d[uniprot].extend([job_title, description, seq])
```
